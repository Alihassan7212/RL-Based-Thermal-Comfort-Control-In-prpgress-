```python
# ==========================================================
# PPO vs Baselines across 4 scenarios (BOPTEST BESTEST HP)
# Scenarios:
#   S1 Morning warm-up (default schedule)
#   S2 Demand Response window (2h cap + cost spike)
#   S3 Capacity derate (constant 60% cap)
#   S4 Noisy sensor (Tin noise) + 1-step actuator lag
# Each scenario runs PPO and non-RL baselines and reports:
#   Return, Energy_kWh, Comfort_degC_h, Smooth_SumAbsDu, Peak_kW, (Cost if DR)
# Also produces a temp gray-band + energy plot for 1 ep per scenario
# ==========================================================

#

import os, sys, time, random
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import BaseCallback

# -------------------------
# Import base env
# -------------------------
MODULE_DIR = "/content/boptestGymService"
MODULE_FILE = os.path.join(MODULE_DIR, "boptestGymEnv.py")
assert os.path.exists(MODULE_FILE), f"Can't find {MODULE_FILE}. Upload or adjust the path."
if MODULE_DIR not in sys.path: sys.path.append(MODULE_DIR)
from boptestGymEnv import BoptestGymEnv as BOPTESTEnvClass
print("Imported: BoptestGymEnv (base)")

# -------------------------
# Global config
# -------------------------

DAY         = 24*3600
STEP_SEC    = 3600          # 1-hour control step
WARMUP      = 1*DAY
EPISODE_LEN = 2*DAY         # 2 days / episode → 48 steps
FIXED_START = 20*DAY        # winter start to ensure load
EXCLUDING   = [(79*DAY, 355*DAY)]

# Reward & comfort
TSET_C, DEADBAND_C = 22.5, 0.5
ETA, ZETA, LAMB = 20.0, 0.8, 0.05   # stronger comfort

# -------------------------
# Utility
# -------------------------
def K_to_C(Tk): return float(Tk) - 273.15

class EpisodeInfo(gym.Wrapper):
    """Accumulate custom episode stats → info['episode_info'] (safe with Monitor)."""
    def __init__(self, env):
        super().__init__(env); self.R=self.E=self.C=self.A=0.0; self.L=0
    def reset(self, **kw): self.R=self.E=self.C=self.A=0.0; self.L=0; return super().reset(**kw)
    def step(self, a):
        obs, rew, term, trunc, info = self.env.step(a)
        self.R += float(rew); self.L += 1
        if isinstance(info, dict) and "reward_terms" in info:
            rt = info["reward_terms"]; self.E += float(rt.get("energy_kWh",0)); self.C += float(rt.get("comfort",0)); self.A += float(rt.get("action_change",0))
        if term or trunc:
            info = dict(info) if isinstance(info, dict) else {}
            info["episode_info"]={"r":self.R,"l":self.L,"energy_kWh":self.E,"comfort":self.C,"action_change":self.A}
        return obs, rew, term, trunc, info

class IndexReward(gym.Wrapper):
    """
    Base env obs order (vector):
      [TinK, P_hp_W, P_fan_W, P_pump_W, ToutK]
    Reward:
      R = -( ETA * comfort + ZETA * step_kWh + LAMB * |Δu| )
      comfort = (dev^2) + 4*max(dev-1,0)^2, dev=max(0, |Tin-Set|-deadband)
    """
    def __init__(self, env, tset_c=TSET_C, deadband_c=DEADBAND_C,
                 eta=ETA, zeta=ZETA, lamb=LAMB, step_seconds=STEP_SEC):
        super().__init__(env)
        self.tset=float(tset_c); self.dead=float(deadband_c)
        self.eta=float(eta); self.zeta=float(zeta); self.lamb=float(lamb)
        self.step_s=int(step_seconds); self.prev_u=None
    def _comfort(self, TinC):
        dev=max(0.0, abs(TinC-self.tset)-self.dead)
        return dev*dev + 4.0*max(0.0, dev-1.0)**2
    def reset(self, **kw): self.prev_u=None; return self.env.reset(**kw)
    def step(self, action):
        obs, _, term, trunc, info = self.env.step(action)
        arr=np.array(obs, float).ravel()
        TinC=K_to_C(arr[0]); P_W=arr[1]+arr[2]+arr[3]
        E_kWh = P_W*(self.step_s/3_600_000.0)
        a = float(action) if np.ndim(action)==0 else float(np.array(action).ravel()[0])
        A = 0.0 if self.prev_u is None else abs(a-self.prev_u); self.prev_u=a
        C = self._comfort(TinC)
        rew = -(self.eta*C + self.zeta*E_kWh + self.lamb*A)
        info = dict(info) if isinstance(info, dict) else {}
        info["reward_terms"]={"energy_kWh":E_kWh,"comfort":C,"action_change":A}
        return obs, rew, term, trunc, info

class IndexObs(gym.Wrapper):
    """Expose to PPO: [Tin°C, P kW, prev_u, Tout°C, sinHr, cosHr]"""
    def __init__(self, env):
        super().__init__(env)
        self.prev_u=0.0; self.t=0
        self.observation_space = gym.spaces.Box(
            low=np.array([-50.0,0.0,0.0,-50.0,-1.0,-1.0], np.float32),
            high=np.array([ 60.0,50.0,1.0,  50.0, 1.0, 1.0], np.float32),
            dtype=np.float32)
        self.action_space = env.action_space
    def _hour_features(self):
        hour = (self.t % 24); theta = 2*np.pi*(hour/24.0)
        return np.sin(theta), np.cos(theta)
    def reset(self, **kw):
        self.prev_u=0.0; self.t=0
        obs, info = self.env.reset(**kw)
        return self._build(obs), info
    def step(self, action):
        obs, rew, term, trunc, info = self.env.step(action)
        a = float(action) if np.ndim(action)==0 else float(np.array(action).ravel()[0])
        self.prev_u=a; self.t += 1
        return self._build(obs), rew, term, trunc, info
    def _build(self, vec):
        arr=np.array(vec, float).ravel()
        TinC=K_to_C(arr[0]); PkW=(arr[1]+arr[2]+arr[3])/1000.0; ToutC=K_to_C(arr[4])
        s,c=self._hour_features()
        return np.array([TinC, PkW, self.prev_u, ToutC, s, c], np.float32)

# ---- scenario wrappers ----
class CapacityCap(gym.Wrapper):
    """Clip action to u_max across all times (e.g., 60% capacity derate)."""
    def __init__(self, env, u_max=0.6): super().__init__(env); self.u_max=float(u_max)
    def step(self, action):
        a=float(np.array(action).ravel()[0]); a_eff=np.array([min(a, self.u_max)], np.float32)
        return self.env.step(a_eff)

class DemandWindowCapAndCost(gym.Wrapper):
    """Between hours [start,end) each day: cap action & add cost spike."""
    def __init__(self, env, start_hour=17, end_hour=19, cap=0.2, p_base=0.10, p_spike=0.60):
        super().__init__(env)
        self.start, self.end = int(start_hour), int(end_hour)
        self.cap=float(cap); self.p_base=float(p_base); self.p_spike=float(p_spike)
        self.t=0
    def reset(self, **kw): self.t=0; return self.env.reset(**kw)
    def step(self, action):
        hour = (self.t % 24); self.t += 1
        in_event = (self.start <= hour < self.end)
        a=float(np.array(action).ravel()[0])
        a_eff=np.array([min(a, self.cap)], np.float32) if in_event else np.array([a], np.float32)
        obs, rew, term, trunc, info = self.env.step(a_eff)
        # compute step cost post-hoc (from reward_terms energy)
        price = self.p_spike if in_event else self.p_base
        rt = info.get("reward_terms", {})
        step_cost = float(rt.get("energy_kWh", 0.0))*price
        info = dict(info) if isinstance(info, dict) else {}
        info.setdefault("scenario_terms", {})["cost_$"] = step_cost
        info["scenario_terms"]["price_$/kWh"] = price
        info["scenario_terms"]["in_event"] = in_event
        return obs, rew, term, trunc, info

class NoisyTinObs(gym.ObservationWrapper):
    """Add noise to Tin°C (first feature of IndexObs output)"""
    def __init__(self, env, sigma=0.2): super().__init__(env); self.sigma=float(sigma)
    def observation(self, obs):
        obs = np.array(obs, np.float32).copy()
        obs[0] = float(obs[0] + np.random.normal(0.0, self.sigma))
        return obs

class ActuatorLag(gym.Wrapper):
    """First-order lag on action: u_applied = alpha*u_prev + (1-alpha)*u_cmd"""
    def __init__(self, env, alpha=0.7):
        super().__init__(env); self.alpha=float(alpha); self.u_prev=0.0
    def reset(self, **kw): self.u_prev=0.0; return self.env.reset(**kw)
    def step(self, action):
        u = float(np.array(action).ravel()[0])
        u_eff = self.alpha*self.u_prev + (1.0-self.alpha)*u
        self.u_prev = u_eff
        return self.env.step(np.array([u_eff], np.float32))

# -------------------------
# Env builder (base) + scenario applicator
# -------------------------
def build_base_env(start_day=FIXED_START):
    observations = {
        "reaTZon_y":    (21+273.15, 24+273.15),   # K
        "reaPHeaPum_y": (0, 6000),                # W
        "reaPFan_y":    (0, 1500),                # W
        "reaPPumEmi_y": (0, 1500),                # W
        "weaSta_reaWeaTDryBul_y": (-30+273.15, 45+273.15),  # outdoor K
    }
    env = BOPTESTEnvClass(
        url=url, testcase=testcase,
        actions=["oveHeaPumY_u"],
        observations=observations,
        random_start_time=False,
        start_time=int(start_day),
        excluding_periods=EXCLUDING,
        max_episode_length=int(EPISODE_LEN),
        warmup_period=int(WARMUP),
        step_period=int(STEP_SEC),
        render_episodes=False
    )
    env = IndexReward(env); env = IndexObs(env); env = EpisodeInfo(env)
    return Monitor(env)

def apply_wrappers(env, wrappers):
    for w, kw in (wrappers or []):
        env = w(env, **(kw or {}))
    return env

# -------------------------
# PPO Training (short for demo)
# -------------------------
class TextProgress(BaseCallback):
    def __init__(self, total, every=200): super().__init__(); self.total=total; self.every=every; self.last=0; self.t0=None
    def _on_training_start(self): self.t0=time.time()
    def _on_step(self)->bool:
        n=self.num_timesteps
        if n-self.last>=self.every:
            sps=n/max(1e-6, time.time()-self.t0)
            print(f"[progress] {n}/{self.total} steps  ~{sps:.1f} steps/s")
            self.last=n
        return True

SEED=123456
np.random.seed(SEED); random.seed(SEED)

train_env = build_base_env()
eval_env  = build_base_env()

STEPS_PER_EP = int(EPISODE_LEN // STEP_SEC)  # 48
EPISODES_TRAIN = 80                           # make bigger (e.g., 150+) if you have time
TIMESTEPS = STEPS_PER_EP * EPISODES_TRAIN

model = PPO("MlpPolicy", train_env,
            learning_rate=3e-4,
            n_steps=STEPS_PER_EP, batch_size=STEPS_PER_EP,
            gamma=0.995, gae_lambda=0.96, ent_coef=0.0,
            clip_range=0.2, verbose=1, seed=SEED)
model.learn(total_timesteps=TIMESTEPS, callback=TextProgress(TIMESTEPS, every=200))

# -------------------------
# Baselines
# -------------------------
def make_hysteresis(delta):
    state={"u":0.0}
    def pol(obs):
        Tin=float(obs[0])
        if Tin < TSET_C - delta: state["u"]=1.0
        elif Tin > TSET_C + delta: state["u"]=0.0
        return state["u"]
    pol.__name__=f"Hysteresis(±{delta})"; return pol

def make_P(kp=0.25):
    state={"u":0.0}
    def pol(obs):
        Tin=float(obs[0])
        if Tin < TSET_C - DEADBAND_C:
            u = float(np.clip(kp*(TSET_C-Tin),0.0,1.0))
        elif Tin > TSET_C + DEADBAND_C:
            u = 0.0
        else:
            u = state["u"]
        state["u"]=u; return u
    pol.__name__=f"P(kp={kp})"; return pol

def make_PID(kp,ki,kd=0.0):
    dt=STEP_SEC/3600.0; state={"u":0.0,"i":0.0,"e_prev":0.0}
    def pol(obs):
        Tin=float(obs[0]); e=(TSET_C-Tin)
        state["i"] += e*dt; d=(e-state["e_prev"])/dt
        u = kp*e + ki*state["i"] + kd*d
        if Tin > TSET_C + DEADBAND_C: u=0.0
        u=float(np.clip(u,0.0,1.0)); state["u"]=u; state["e_prev"]=e; return u
    pol.__name__=f"PID(kp={kp},ki={ki},kd={kd})"; return pol

def make_const(u): f=lambda obs: float(u); f.__name__=f"Const({u})"; return f

# -------------------------
# Evaluation helpers
# -------------------------
def run_episode(env, controller, is_rl=False):
    obs,_ = env.reset(); done=False
    R=E=C=A=0.0; peak=0.0; cost=0.0
    while not done:
        if is_rl:
            act,_=model.predict(obs, deterministic=True)
        else:
            u=float(controller(obs)); act=np.array([u], np.float32)
        obs, rew, term, trunc, info = env.step(act)
        done=bool(term or trunc)
        R += float(rew); peak=max(peak, float(obs[1]))
        rt=info.get("reward_terms", {})
        E += float(rt.get("energy_kWh",0.0))
        C += float(rt.get("comfort",0.0))
        A += float(rt.get("action_change",0.0))
        st=info.get("scenario_terms", {})
        cost += float(st.get("cost_$",0.0))
    try: env.unwrapped.stop()
    except: pass
    return {
        "Return": R,
        "Energy_kWh": E,
        "Comfort_degC_h": C,
        "Smooth_SumAbsDu": A,
        "Peak_kW": peak,
        "Cost_$": cost,
    }


def eval_policy(make_env_fn, controller, episodes=5, name=""):
    rows=[]
    for ep in range(episodes):
        env = make_env_fn()
        rows.append(run_episode(env, controller, is_rl=(controller=="RL") ))
    df = pd.DataFrame(rows); df.insert(0, "Policy", name)
    return df

def summarize(df):
    g = df.groupby("Policy").agg(["mean","std"])
    g.columns = [f"{c[0]}_{c[1]}" for c in g.columns]
    return g.reset_index()

# -------------------------
# Scenario definitions
# -------------------------
def make_env_S1():  # Morning warm-up (default)
    return build_base_env()

def make_env_S2():  # DR window: 17-19h cap to 0.2 + price spike
    env = build_base_env()
    env = DemandWindowCapAndCost(env, start_hour=17, end_hour=19, cap=0.2, p_base=0.10, p_spike=0.60)
    return env

def make_env_S3():  # Constant 60% capacity
    return CapacityCap(build_base_env(), u_max=0.6)

def make_env_S4():  # Noisy sensor + actuator lag
    env = build_base_env()
    env = NoisyTinObs(env, sigma=0.2)
    env = ActuatorLag(env, alpha=0.7)
    return env

SCENARIOS = [
    ("S1: Morning warm-up", make_env_S1, dict(show_cost=False)),
    ("S2: DR window (cap+price)", make_env_S2, dict(show_cost=True)),
    ("S3: Constant 60% derate", make_env_S3, dict(show_cost=False)),
    ("S4: Noisy Tin + lag", make_env_S4, dict(show_cost=False)),
]

# -------------------------
# Run all scenarios
# -------------------------
EP_BENCH = 5  # per-scenario episodes (increase for tighter stats)
all_results = {}

for scen_name, scen_env_fn, opts in SCENARIOS:
    print(f"\n===== {scen_name} =====")
    # quick tuning for fair baselines (small grid)
    comfort_cap = 60.0  # acceptable comfort budget; tweak per scenario if needed

    # hysteresis tuning
    best_hys=None; best_row=None
    for d in [0.3,0.5,0.8,1.0,1.2]:
        df=eval_policy(scen_env_fn, make_hysteresis(d), episodes=3, name=f"Hysteresis(±{d})")
        row=df.mean(numeric_only=True); row["name"]=f"Hysteresis(±{d})"
        if row["Comfort_degC_h"] <= comfort_cap and (best_row is None or row["Energy_kWh"] < best_row["Energy_kWh"]):
            best_row=row; best_hys=make_hysteresis(d)
    if best_hys is None: best_hys=make_hysteresis(1.0)

    # PID tuning
    best_pid_fn=None; best_pid=None
    for kp in [0.2,0.3,0.4]:
        for ki in [0.02,0.05]:
            df=eval_policy(scen_env_fn, make_PID(kp,ki,0.0), episodes=3, name=f"PID({kp},{ki})")
            row=df.mean(numeric_only=True); row["name"]=f"PID({kp},{ki})"
            if row["Comfort_degC_h"] <= comfort_cap and (best_pid is None or row["Energy_kWh"] < best_pid["Energy_kWh"]):
                best_pid=row; best_pid_fn=make_PID(kp,ki,0.0)
    if best_pid_fn is None: best_pid_fn=make_PID(0.3,0.05,0.0)

    # final benchmark
    dfs = [
        eval_policy(scen_env_fn, "RL", episodes=EP_BENCH, name="PPO-RL"),
        eval_policy(scen_env_fn, best_hys, episodes=EP_BENCH, name=getattr(best_hys,"__name__","Hysteresis")),
        eval_policy(scen_env_fn, best_pid_fn, episodes=EP_BENCH, name=getattr(best_pid_fn,"__name__","PID")),
        eval_policy(scen_env_fn, make_P(0.25), episodes=EP_BENCH, name="P-only"),
        eval_policy(scen_env_fn, make_const(0.0), episodes=EP_BENCH, name="Const(0)"),
        eval_policy(scen_env_fn, make_const(0.5), episodes=EP_BENCH, name="Const(0.5)"),
    ]
    results = pd.concat(dfs, ignore_index=True)
    all_results[scen_name] = results

    print("Per-episode results:"); display(results)
    print("\nSummary (mean ± std):"); display(summarize(results))

    # quick bars (Return, Energy, Comfort, Peak; add Cost if scenario 2)
    metrics = ["Return","Energy_kWh","Comfort_degC_h","Peak_kW"]
    if opts.get("show_cost"): metrics.append("Cost_$")
    fig, axes = plt.subplots(1, len(metrics), figsize=(4*len(metrics), 4))
    for ax, m in zip(np.atleast_1d(axes), metrics):
        s = results.groupby("Policy")[m].mean().sort_values(ascending=(m!="Return"))
        s.plot(kind="bar", ax=ax)
        ax.set_title(f"{m} ({scen_name.split(':')[0]})"); ax.grid(True, alpha=0.3)
    plt.tight_layout(); plt.show()

    # one nice episode plot (temp band + energy)
    # build a fresh env and roll one deterministic episode
    env = scen_env_fn()
    obs,_ = env.reset()
    TinC=[]; Tlo=[]; Thi=[]; P_hp=[]; P_fan=[]; P_pump=[]; Ptot=[]; EkWh=[]; U=[]; hrs=[]
    have_sp=True
    # tap a plain env exposing setpoints for plotting
    try:
        obs_plot = {
            "reaTZon_y": (21+273.15,24+273.15),
            "reaTSetHea_y": (10+273.15,26+273.15),
            "reaTSetCoo_y": (20+273.15,35+273.15),
            "reaPHeaPum_y": (0,6000), "reaPFan_y": (0,1500), "reaPPumEmi_y": (0,1500),
            "weaSta_reaWeaTDryBul_y": (-30+273.15,45+273.15)
        }
        tap = BOPTESTEnvClass(url=url, testcase=testcase, actions=["oveHeaPumY_u"],
                              observations=obs_plot, random_start_time=False,
                              start_time=int(FIXED_START), excluding_periods=EXCLUDING,
                              max_episode_length=int(EPISODE_LEN), warmup_period=int(WARMUP),
                              step_period=int(STEP_SEC), render_episodes=False)
    except Exception:
        have_sp=False
    t=0; done=False
    while not done:
        act,_=model.predict(obs, deterministic=True)
        obs,_,tw,uw,info = env.step(act); done=bool(tw or uw)
        if have_sp:
            obsp,_,tp,up,_ = tap.step(act)
            arr=np.array(obsp, float).ravel(); TinC.append(K_to_C(arr[0])); Tlo.append(K_to_C(arr[1])); Thi.append(K_to_C(arr[2]))
            hp,fan,pump = arr[3],arr[4],arr[5]
        else:
            # fallback: get Tin and power from IndexObs features via env info
            TinC.append(float(obs[0])); Tlo.append(TSET_C-DEADBAND_C); Thi.append(TSET_C+DEADBAND_C)
            # power feature in IndexObs already kW at obs[1]
            hp=fan=pump=0.0
        P_hp.append(hp/1000.0); P_fan.append(fan/1000.0); P_pump.append(pump/1000.0)
        Ptot.append((hp+fan+pump)/1000.0)
        rt=info.get("reward_terms", {}); EkWh.append(float(rt.get("energy_kWh",0.0)))
        U.append(float(np.array(act).ravel()[0])); hrs.append(t); t+=1
    try: env.unwrapped.stop()
    except: pass
    try: tap.unwrapped.stop()
    except: pass

    # plot
    hrs = np.array(hrs)
    plt.figure(figsize=(10,9))
    ax1=plt.subplot(3,1,1)
    plt.fill_between(hrs, Tlo, Thi, step='post', alpha=0.25, color='gray', label='Comfort band')
    plt.step(hrs, TinC, where='post', linewidth=2, label='Zone °C')
    ax1.set_ylabel("Temp (°C)"); ax1.set_title(f"{scen_name}: temperature vs band"); ax1.grid(True, alpha=0.3); ax1.legend()

    ax2=plt.subplot(3,1,2)
    if any(P_hp):
        ax2.stackplot(hrs, P_hp, P_fan, P_pump, labels=["HP kW","Fan kW","Pump kW"], step='post', alpha=0.7)
        ax2.step(hrs, Ptot, where='post', linewidth=2, label="Total kW")
    else:
        ax2.step(hrs, np.array(EkWh)*(3600.0/STEP_SEC), where='post', linewidth=2, label="Approx kW")
    ax2.set_ylabel("Power (kW)"); ax2.grid(True, alpha=0.3)
    ax2b=ax2.twinx(); ax2b.step(hrs, U, where='post', linestyle='--', linewidth=1.5, label="u (action)")
    ax2b.set_ylabel("Action"); ax2b.set_ylim(-0.05,1.05)
    h1,l1=ax2.get_legend_handles_labels(); h2,l2=ax2b.get_legend_handles_labels()
    ax2b.legend(h1+h2, l1+l2, loc='upper right')

    ax3=plt.subplot(3,1,3)
    plt.step(hrs, np.cumsum(EkWh), where='post', linewidth=2)
    ax3.set_xlabel("Hour"); ax3.set_ylabel("Cumulative kWh"); ax3.grid(True, alpha=0.3)
    ax3.set_title(f"{scen_name}: cumulative energy")
    plt.tight_layout(); plt.show()

print("\nAll scenarios complete.")
```
